```python
import pandas as pd
import requests
from io import StringIO

url = "https://drive.google.com/uc?export=download&id=1xxDtrZKfuWQfl-6KA9XEd_eatitNPnkB"

# Fetch the content of the CSV file without verifying SSL certificate
response = requests.get(url, verify=False)

# Read CSV content into a DataFrame
df = pd.read_csv(StringIO(response.text))
```

```python
df
```

```python
# split data
```

```python
X = df.drop('price', axis = 1)
y = df['price']

print('Shape of X = ', X.shape)
print('Shape of y = ', y.shape)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2, random_state=51)
print('Shape of X_train = ', X_train.shape)
print('Shape of X_test = ', X_test.shape)
print('Shape of y_train = ', y_train.shape)
print('Shape of y_test = ', y_test.shape)
```

```python
# feature scalling
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
sc.fit(X_train)
X_train = sc.transform(X_train)
X_test = sc.transform(X_test)
```

```python
#Linear Regression - ML Model Training
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()

lr.fit(X_train, y_train)
```

```python
lr.coef_
```

```python
lr.intercept_
```

```python
# Predict the value of home and test
```

```python
X_test[0,:]
```

```python
lr.predict([X_test[0,:]])
```

```python
lr.predict(X_test)
```

```python
y_test
```

```python
# per of accuracy
```

```python
lr.score(X_test, y_test)
```

```python
# Ridge and Lasso regression - over fit model to generalise
```

```python
from sklearn.linear_model import Ridge, Lasso
```

```python
rd = Ridge()

rd.fit(X_train, y_train)
```

```python
rd.score(X_train, y_train)
```

```python
ls = Lasso()

ls.fit(X_train, y_train)
ls.score(X_train, y_train)
```

```python
rd2 = Ridge(alpha = 2)

rd2.fit(X_train, y_train)
rd2.score(X_train, y_train)
```

```python
ls2 = Lasso(alpha = 3)

ls2.fit(X_train, y_train)
ls2.score(X_train, y_train)
```

```python
# Root Mean square error
```

```python
# Model evaluatio
```

```python
y_pred = lr.predict(X_test)
y_pred
```

```python
y_test
```

```python
from sklearn.metrics import mean_squared_error
import numpy as np
```

```python
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)

print('Mse =',  mse)
print('Rmse =',  rmse)
```

```python
# R-squared regression
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_test, y_pred)
```

```python
# Polynomial Regression Model - Ml Model training
```
