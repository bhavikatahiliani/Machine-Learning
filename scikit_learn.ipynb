```python
train.head()
```

```python
x_train = train.drop(columns = "price")
y_train = train["price"]
```

```python
#numarical missing value imputation
```

```python
num_vars = x_train.select_dtypes(include=["int64", "float64"]).columns
```

```python
num_vars
```

```python
x_train[num_vars].isnull().sum()
```

```python
imputer_mean = SimpleImputer(strategy = "mean")
```

```python
imputer_mean.fit(x_train[num_vars]) #num value's mean
```

```python
imputer_mean.transform(x_train[num_vars]) # always gives array
```
